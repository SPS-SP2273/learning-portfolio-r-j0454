```python
import os
import glob
import shutil
```

## Task 1 & 2

```python
os.getcwd()

#Specifying location of data folder:
data_folder = '/Users/rijul/Downloads/CODING/os-collaboration-exercise-data'


txt_files = glob.glob(os.path.join(data_folder, "*.txt"))     # *.txt tells python to only join files with extension .txt to the data_folder:)
```

```python
# Define function to convert filename:
def convert_filename(filename):
    
    name, extension = filename.split('.')
    name = filename.replace(".txt", "")

    #if file already renamed:
    if name[0:2].isdigit() and name[3:5].isdigit() and name[2] == "-":
        return filename
    
    else: 
        month_dict = {
        'jan': '01', 'feb': '02', 'mar': '03',
        'apr': '04', 'may': '05', 'jun': '06',
        'jul': '07', 'aug': '08', 'sep': '09',
        'oct': '10', 'nov': '11', 'dec': '12'
        }

        #Split filename
        month_day, city = name.split('_')
        month, day = month_day.split('-')

        #Convert month
        month_number = month_dict[month.lower()]

        #change filename:
        new_filename = f"{month_number}-{day}_{city}.{extension}"
        return new_filename


convert_filename("oct-08_singapore.txt")

```

## Task 3

```python
# Using a for loop to convert following list of filenames:
file_name_list = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

for filename in file_name_list:
    print(f'{filename} ---> {convert_filename(filename)}')
```

## Task 4

```python
#Use glob to rename all files in the folder

for old_path in txt_files:
    old_file_name = os.path.basename(old_path)
    new_file_name = convert_filename(old_file_name)
    
    # Skip if already renamed
    if old_file_name == new_file_name:
        print(f"Skipped (already renamed): {old_file_name}")
        continue
    
    new_path = os.path.join(os.path.dirname(old_path), new_file_name)
    
    if os.path.exists(old_path):   # only rename if the file exists
        os.rename(old_path, new_path)
        print(f"Renamed: {old_file_name}  {new_file_name}")
```

## Task 5

```python

cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

cityfolder_path = '/Users/rijul/Downloads/CODING/os-collaboration-exercise-data'

# For every city, create a path using Path.join
for city in cities:
    folder_path = os.path.join(cityfolder_path, city.lower())

# And for each of those paths make a folder/directory, and say created folder to indicate completion
    os.mkdir(folder_path)
    print(f"Created folder: {folder_path}")
```

## Task 6

```python
# Path of the folder named 'cairo' in data_folder as a variable:
cairo_folder = os.path.join(data_folder, 'cairo')

# Selects all files that have 'cairo.txt'. in them
cairo_files = glob.glob(os.path.join(data_folder, '*cairo.txt'))

print(cairo_files)
```

```python
# Makes a copy of each selected file in cairo_files in cairo_folder
for file_path in cairo_files:
    shutil.copy(file_path, cairo_folder)
    print(f'Copied {file_path} ----> {cairo_folder}')
```

## Task 7

```python
# Now for all cities in cities:

for city in cities:
    # Destination city folders defined
    city_folder = os.path.join(data_folder, city)

    # find all files with corresponding city name
    city_files = glob.glob(os.path.join(data_folder, f'*{city.lower()}.txt'))

    # makes copies of each file path in city_files in the corresponding city_folder
    for file_path in city_files:
        shutil.copy(file_path, city_folder)
        print(f'Copied {file_path} ----> {city_folder}')


```
