## Exercise 1: Celsius to Fahrenheit or Kelvin

```python
def convert_celsius():
    temperature_celsius = float(input('Temperature in ˚C: '))
    target_scale = input('Scale to convert to (F or K): ')
    if target_scale.upper() == 'K':
        temp_convert = temperature_celsius + 273.15
        print(f'{temp_convert:.2f} K')
    else:
        temp_convert = (9/5)*temperature_celsius + 32
        print(f'{temp_convert:.2f} ˚F')

convert_celsius()
```

## Exercise 2: Fahrenheit to Celsius or Kelvin:

```python
def convert_fahrenheit():
    temperature_fahrenheit = float(input('Temperature in ˚F: '))
    target_scale = input('Scale to convert to (C or K): ')
    if target_scale.upper() == 'K':
        temp_convert = (temperature_fahrenheit-32)*(5/9) + 273.15
        print(f'{temp_convert:.2f} K')
    else:
        temp_convert = (temperature_fahrenheit-32)*(5/9)
        print(f'{temp_convert:.2f} ˚C')

convert_fahrenheit()
```

## Exercise 3: General temperature converter

```python
def convert_temperature():
    source_scale = input('Scale to convert FROM (C, F, or K): ')
   
    if source_scale.upper() == 'K':
        temperature_kelvin = float(input('Temperature in K: '))
        target_scale = input('Scale to convert to (F or C): ')
        if target_scale.upper() == 'C':
            temp_convert = temperature_kelvin - 273.15
            print(f'{temp_convert:.2f} ˚C')
        else:
            temp_convert = (9/5)*(temperature_kelvin-273.15) + 32
            print(f'{temp_convert:.2f} ˚F')
    
    elif source_scale.upper() == 'C':
        convert_celsius()

    elif source_scale.upper() == 'F':
        convert_fahrenheit()


convert_temperature()
        



```
