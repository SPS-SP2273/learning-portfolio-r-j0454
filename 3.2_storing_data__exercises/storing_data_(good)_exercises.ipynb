## Exercise 2: Show me the 'odd' letters

```python
np_array_2d = np.array([[1, "A"], [3, "C"], [2, "B"], [4, "D"],
                        [5, "E"], [7, "G"], [6, "F"], [8, "H"],
                        [10, "J"], [9, "I"]])


np_array_2d[:,0]
```

```python
#array_name.astype() in numpy converts from one type of data to another (e.g. str to int)--->"array_name.astype()=CHANGE THE TYPE"
first_column_int = np_array_2d[:,0].astype(int)            #converted the first entry column into an array of integers so we can do operations and mask.
```

```python
remainder = first_column_int % 2
remainder
```

```python
mask = first_column_int % 2 == 1
mask
```

```python
np_array_2d[mask]       #only outputs rows where remainder of first_column_int/2 is 1 (odd rows)
```
