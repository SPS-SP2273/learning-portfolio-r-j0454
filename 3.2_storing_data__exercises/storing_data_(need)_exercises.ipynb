## Exercise 5: A Matter of Statistics

```python
x_1 = np.array([1, 4, 9, 16, 25, 36, 49, 64, 81, 100])
x_2 = np.array([1, 8, 27, 64, 125, 216, 343, 512, 729, 1000])
```

```python
# Mean of x_1 and x_2:
print(np.mean(x_1))
print(np.mean(x_2))

```

```python
#Standard deviations:
print(np.std(x_1))
print(np.std(x_2))
```

For reference:
$r = \frac{\sum_{i=1}^{n}(x_i-\mu\chi)(y_i-\mu\Upsilon)}{\sqrt{\sum_{i=1}^{n}(x_i-\mu\chi)^2\sum_{i=1}^{n}(y_i-\mu\Upsilon)^2}}$

```python
#Manually calculating r:
mean_1 = np.mean(x_1)
mean_2 = np.mean(x_2)
r = (np.sum((x_1-mean_1)*(x_2-mean_2)))/(np.sqrt(np.sum((x_1-mean_1)**2))*np.sqrt(np.sum((x_2-mean_2)**2)))
print(r)
```

```python
#using np.corrcoef:
np.corrcoef(x_1, x_2)
```

Very close results as we can see, though np.corrcoef() gives it in this interesting array rather than a single number
