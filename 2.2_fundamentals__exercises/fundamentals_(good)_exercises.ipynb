## Exercise 2: What is your grade?

````python
grade = input("Enter the student's score: ")
grade = float(grade)

if 70<=grade<=100:
    print("The student's letter grade is: A")
elif 50<=grade<=69:
    print("The student's letter grade is: B")
elif 35<=grade<=49:
    print("The student's letter grade is: C")
else:
    print("The student has failed :(")
````

## Exercise 3: Debug code :))

````python
def factorial(n):
    if n < 0:
        return "Invalid input"
    elif n != 0:
        result = 1
        for number in range(1, n+1):
            result *= number
        return result
    else:
        return 1


print(factorial(5))   # Testing, expected output: 120
print(factorial(-1))  # Testing, expected output: Invalid input
````

errors were:
1. first line should be n<0 not n>0 
2. the result code should be under 'elif n!=0:' not under 'else:'
3. 'result' should be:
```python
'result *= number' 
```
NOT
```python
'result += number' 
```
in order to correctly compute a factorial.
