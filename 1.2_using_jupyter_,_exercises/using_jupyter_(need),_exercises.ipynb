## 1 Quadratic Equations


### Introduction
![](https://upload.wikimedia.org/wikipedia/commons/thumb/f/f8/Polynomialdeg2.svg/300px-Polynomialdeg2.svg.png)


The general form of a quadratic equation is:

$$ax^2 + bx + c = 0$$

### Solutions

Provided $a≠0$, we can use an algebraic method called [completing the square](https://en.wikipedia.org/wiki/Completing_the_square) to show that a quadratic equation has the following solution:

$$x = \frac{-b±\sqrt{b^2-4ac}}{2a}$$


### Discriminant

The quantity $\Delta = b^2-4ac$ is called the **discriminant** of the equation and decides the nature of its solutions. The table below shows the various possibilities. 

| **Discriminant** | **Roots** |
|------------------|-----------|
|$\Delta = b^2-4ac = 0$| A single solution of $-b/(2a)$|
|$\Delta = b^2-4ac > 0$| Two distinct real solutions |
|$\Delta = b^2-4ac < 0$| No real solutions; both are complex.|
