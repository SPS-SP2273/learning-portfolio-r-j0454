# 1 The for iterator

Say you want to print a message corresponding to every superhero in the following list:

```python
real_names = ["Natasha Romanoff", "Tony Stark", "Stephen Strange"]
```

```python
#BAD way to do this:
name=real_names[0]
print(f"{name} is a Marvel superhero!")

name=real_names[1]
print(f"{name} is a Marvel superhero!")

name=real_names[2]
print(f"{name} is a Marvel superhero!")
```

The above method is WAY too tedious. 

Instead of the above, we use 'for' loops:

## 1.1 for with a list

```python
for name in real_names:
    print(f"{name} is a Marvel superhero!")
```

## 1.2 for with enumerate

```python
super_names = ["Black Widow", "Iron Man", "Doctor Strange"]
real_names = ["Natasha Romanoff", "Tony Stark", "Stephen Strange"]
```

Since the for loop only accepts one list, we need to do something else to access the data in both lists. One option is to use enumerate(). Let's see how enumerate() works.

```python
for count, name in enumerate(real_names):
    print(f'{count}: {name} is a Marvel superhero!')
```

enumerate() basically keeps count. In the above example, enumerate() not only gives the elements of the list, it also gives you a number (that is stored in count).

```python
for index, name in enumerate(real_names):
    superhero_name = super_names[index]
    print(f'{name} is {superhero_name}!')
```

## 1.3 for with range

```python
for i in range(0,11):
    print(i)
```

```python
for i in range(len(real_names)):
    real_name = real_names[i]
    super_name = super_names[i]        
    print(f"{real_name} is Marvel's {super_name}!")
```
