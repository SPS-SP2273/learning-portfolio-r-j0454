## Exercise 1: Make me an odd list (For loop)

```python
list=[]
for i in range(10):
    if i%2==0:
        continue
    else:
        list.append(i**2)
list
    
```

## Exercise 2: Make me another odd list (list comprehension)

```python
[i**2 for i in range(10) if i%2 != 0]
```

## Exercise 3: Time me!

```python
%%timeit
list=[]
for i in range(10):
    if i%2==0:
        continue
    else:
        list.append(i**2)
list
```

```python
%%timeit
[i**2 for i in range(10) if i%2 != 0]
```

**List comprehension takes more time than for loop**---->**for loop is faster**

## Exercise 4: A problem of decay

```python
sample_0 = 100     #Initial amount of substance at the beginning, sample naught
years = 0
sample_n = 100     #What was current amount of sample right now? (100 when years = 0)

while sample_n >= 0.5 * sample_0:       #iterates until sample_n is half of sample_0, while adding to 'years' variable, then output the value of 'years'
    sample_n = sample_n*0.95
    years+=1

print(years)
```

## Exercise 5: Changes in CO<sub>2<sub>

```python
co2_data = [
    (2000, 369.55), (2001, 371.14), (2002, 373.28), 
    (2003, 375.80), (2004, 377.52), (2005, 379.80), 
    (2006, 381.90), (2007, 383.79), (2008, 385.60), 
    (2009, 387.43), (2010, 389.90), (2011, 391.65), 
    (2012, 393.85), (2013, 396.52), (2014, 398.65),
    (2015, 400.83), (2016, 404.24), (2017, 406.55), 
    (2018, 408.52), (2019, 411.44), (2020, 414.24)
]

for (year_prev, co2_prev), (year_curr, co2_current) in zip(co2_data, co2_data[1:]):
    change = co2_current - co2_prev
    if change >= 3:
        print(f'Year: {year_curr}: Increase = {change:.2f} ppm')
```

**Task**: Identify those years that showed an increase of CO
 of 3 ppm or more compared to the previous year.

Please print out these years along with the corresponding change in concentration.
