## Exercise 2: Multiplication Table

Using a nested for loop:

````python
for letter in ['A', 'B', 'C']:
    for number in [1, 2, 3]:
        print(f'{letter}{number}', end='\t')
    print('\n')
````

Note that the 
```python
print(f'', end='\t')  
```
the ```'\t'```instructs the computer to add a tab after the i*j, and not 
```'\n'``` which would be a new line (and which is the default)

````python
for i in range(1,6):
    print(f'{i}:', end='\t')
    for j in range(1,6):
        print(f"{i*j}", end='\t')        
    print('\n\n')
````
