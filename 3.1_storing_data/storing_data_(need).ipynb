# 2 Lists vs. Arrays

## 2.1 Size

````python
py_list_2d = [[1, "A"], [2, "B"], [3, "C"], [4, "D"],
              [5, "E"], [6, "F"], [7, "G"], [8, "H"],
              [9, "I"], [10, "J"]]

np_array_2d = np.array(py_list_2d)      # Reusing the Python list 
                                        # to create a NEW
                                        # NumPy array
````

````python
len(py_list_2d)
len(np_array_2d)
np_array_2d.shape
````

The above (np_array_2d.shape) tells me both the amount of entries in the 2d list (the grouped in [] entries) AND also tells me the dimension of the list (2D in this case so 2):

resulting output of 
```python
np_array_2d.shape
``` 
---> (10, 2)

## 2.2 Arrays are fussy about type

````python
py_list = [1, 1.5, 'A']
np_array = np.array(py_list)

py_list
np_array
````

## 2.3 Adding a number

````python
py_list = [1, 2, 3, 4, 5]
np_array = np.array(py_list)         # Reusing the Python list
                                     # to create a NEW
                                     # NumPy array

np_array + 10 #Adds 10 to all elements
````

````python
py_list + 10        # Won't work! Only works with arrays and numpy arrays
````

## 2.4 Adding another list

````python
py_list_1 = [1, 2, 3, 4, 5]
py_list_2 = [10, 20, 30, 40, 50]

np_array_1 = np.array(py_list_1)
np_array_2 = np.array(py_list_2)
````

````python
py_list_1 + py_list_2
np_array_1 + np_array_2.   #its like vector addition each like index is added to the other's
````

## 2.8 Mathematics

````python
py_list = [1, 2, 3, 4, 5]
np_array = np.array(py_list)         
````

````python
sum(py_list)     # sum() is a base Python function
max(py_list)     # max() is a base Python function
min(py_list)     # min() is a base Python function
np_array.sum()
np_array.max()
np_array.min()
np_array.mean()
np_array.std()
````
